__2. What was the percentage of female basketball players among all the female participants of the 2012 Olympics? Round the answer to the first decimal.__

__8. Is it true that Spain won fewer medals than Italy at the 2016 Olympics? Do not consider NaN values in _Medal_ column.__ 



```python
spain=data[(data['Team']=='Spain')&(data['Year']==2016) &(pd.isna(data['Medal'])==False)]
italy=data[(data['Team']=='Italy')&(data['Year']==2016) &(pd.isna(data['Medal'])==False)]

italy_count = len(italy.index)
spain_count = len(spain.index)
print('Spain won: ', spain_count, 'Medails in 2016')
print('Italy won: ', italy_count, 'Medails in 2016')
if spain_count > italy_count:
    print('Spain won more medals in 2016 than italy')
else:
    print('Italy won more medals than spain in 2016')



```
